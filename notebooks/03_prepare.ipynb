# Prepare

## What you learn

Analysis Reportをステークホルダーに共有し、必要なデータや不足しているデータの収集・作成を行いDatasetを作成します。

## Why you need

Business goalを達成するためには、必要十分なデータを収集・作成する必要があります。

## How to do

Analyzeのプロセスで作成したAnalysis Reportをもとに、データの収集・作成が可能なステークホルダーに協力を依頼します。次の3つを明確にするとよいでしょう。

* Ask
   * どんなデータが、いつまでに、どれくらいほしいか。
* Necessity
   * なぜそのデータが必要なのか。
* How to provide
   * どのようにデータを収集・作成できるか。

### Technical Exercise

データを作成する場合は、専用のツールを使用して自分・チームで作成するか、外注を行います。

データ作成ツールの例

* [doccano](https://github.com/doccano/doccano)
* [Label Studio](https://labelstud.io/)
* [Prodigy](https://prodi.gy/)

### Communication Exercise

* Product Manager

#### Example

* Product Manager
   * データを追加で収集・作成するための予算はあるか。
* Data architect/Software Engineer
   * データの収集にかかる工数はどの程度か。
* Domain Expert
   * 現在どのような基準で価格を設定しているか。マニュアルなどはあるか。
